## Closed Form

```python
from sklearn.datasets import make_regression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=2)
print(x.shape)
x=(x-x.mean())/x.std()
plt.scatter(x,y)
print(y.shape)
```

```python
y=y.reshape((-1,1))
ones=np.ones((400,1))
X=np.hstack((x,ones))
print(X.shape)
```

```python
print(X[2:9])
```

```python
def predict(X,theta):
    return np.dot(X,theta)
def closedform(Y,X):
    Y=np.mat(Y)
    firstPart=np.dot(X.T,X)
    secondPart=np.dot(X.T,Y)
    theta=np.linalg.pinv(firstPart)*secondPart
    return theta
```

```python
theta=closedform(y,X)
print(theta)
```

```python
plt.scatter(x,y)
plt.plot(x,predict(X,theta),color="orange")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X=dfx.values
Y=dfy.values
```

```python
print(X.shape)
```

```python
X=(X-X.mean())/X.std()
plt.style.use("seaborn")
```

```python
plt.scatter(X,Y)
plt.title("Normalised")
```

```python
def getW(X,x_query,t):
    m=X.shape[0]
    W=np.eye(m)
    x=x_query
    i=0
    while (i<m):
        W[i][i]=np.exp((np.dot((X[i]-x),(X[i]-x).T))/(-(2*t*t)))
        i+=1
    return W
```

```python
xq=np.mat([2,1])
W=getW(X,xq,0.2)
```

```python
def predict(X,Y,xq,t):
    X_=np.ones((99,1))
    X_=np.hstack((X,X_))
    X_=np.mat(X_)
    Y=np.mat(Y)
    W=getW(X,xq,t)
    xq=np.mat([xq,1])
    theta = ( np.linalg.pinv(X_.T*(W*X_)) ) * (X_.T*(W*Y))
    pre=np.dot(xq,theta)
    return theta,pre
```

```python
def hypothesis(theta,x):
    x=np.mat(x)
    theta=np.mat(theta)
    z=(np.dot(x,theta))
    return sigmoid(z)
```

```python
def sigmoid(z):
    k=(1.0/(1.0+np.exp(-1.0*z)))
    return k
```

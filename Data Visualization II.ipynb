```python
import numpy as np
import matplotlib.pyplot as plt

u1=np.array([2,4])
cov_1=np.array([[1,0],[0,1]])
u2=np.array([-1,-3])
cov_2=np.array([[1,0],[0,1]])
```

```python
x1=np.random.multivariate_normal(u1,cov_1,300)
x2=np.random.multivariate_normal(u2,cov_2,400)
print(x1.shape)
print(x2.shape)
```

```python
plt.scatter(x1[:,0],x1[:,1])
plt.scatter(x2[:,0],x2[:,1]) #in x1[0] we have sweetness x1[1] we have colour
plt.show()
```
